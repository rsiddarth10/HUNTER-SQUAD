```python
import os
import torch
from PIL import Image
import torchvision.transforms as T

class PotholeDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms

        self.imgs = sorted(os.listdir(os.path.join(root, "images")))
        self.labels = sorted(os.listdir(os.path.join(root, "labels")))

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "images", self.imgs[idx])
        label_path = os.path.join(self.root, "labels", self.labels[idx])

        img = Image.open(img_path).convert("RGB")
        w, h = img.size

        boxes = []
        labels = []

        with open(label_path) as f:
            for line in f.readlines():
                cls, x, y, bw, bh = map(float, line.strip().split())

                # YOLO → pixel conversion
                x1 = (x - bw/2) * w
                y1 = (y - bh/2) * h
                x2 = (x + bw/2) * w
                y2 = (y + bh/2) * h

                boxes.append([x1, y1, x2, y2])
                labels.append(1)  # pothole class = 1

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)

        target = {
            "boxes": boxes,
            "labels": labels
        }

        if self.transforms:
            img = self.transforms(img)

        return img, target
```

```python
def get_transform():
    return T.Compose([
        T.ToTensor()
    ])

train_dataset = PotholeDataset(
    "/content/drive/MyDrive/dataset_final/train",
    get_transform()
)

val_dataset = PotholeDataset(
    "/content/drive/MyDrive/dataset_final/valid",
    get_transform()
)

train_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True,
    collate_fn=lambda x: tuple(zip(*x))
)

val_loader = torch.utils.data.DataLoader(
    val_dataset,
    batch_size=4,
    shuffle=False,
    collate_fn=lambda x: tuple(zip(*x))
)
```

```python
import torch

def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter_area = max(0, x2 - x1) * max(0, y2 - y1)

    box1_area = (box1[2]-box1[0]) * (box1[3]-box1[1])
    box2_area = (box2[2]-box2[0]) * (box2[3]-box2[1])

    union = box1_area + box2_area - inter_area

    return inter_area / union if union > 0 else 0


def calculate_map(model, dataloader, device, iou_threshold=0.5, score_threshold=0.5):
    model.eval()

    TP, FP, FN = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images = list(img.to(device) for img in images)
            outputs = model(images)

            for i in range(len(outputs)):
                preds = outputs[i]
                gt_boxes = targets[i]['boxes']

                pred_boxes = preds['boxes'].cpu()
                scores = preds['scores'].cpu()

                keep = scores > score_threshold
                pred_boxes = pred_boxes[keep]

                matched = set()

                for pb in pred_boxes:
                    best_iou = 0
                    best_gt = -1

                    for j, gb in enumerate(gt_boxes):
                        iou = compute_iou(pb.tolist(), gb.tolist())
                        if iou > best_iou:
                            best_iou = iou
                            best_gt = j

                    if best_iou >= iou_threshold and best_gt not in matched:
                        TP += 1
                        matched.add(best_gt)
                    else:
                        FP += 1

                FN += len(gt_boxes) - len(matched)

    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)
    mAP = precision * recall

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"mAP@0.5: {mAP:.4f}")

    return mAP
```

```python
def evaluate(model, dataloader, device, iou_threshold=0.5, score_threshold=0.5):
    model.eval()

    TP, FP, FN = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images = [img.to(device) for img in images]
            outputs = model(images)

            for i in range(len(outputs)):
                pred_boxes = outputs[i]['boxes'].cpu()
                scores = outputs[i]['scores'].cpu()
                gt_boxes = targets[i]['boxes'].cpu()

                keep = scores > score_threshold
                pred_boxes = pred_boxes[keep]

                matched = set()

                for pb in pred_boxes:
                    best_iou = 0
                    best_gt = -1

                    for j, gb in enumerate(gt_boxes):
                        iou = compute_iou(pb.tolist(), gb.tolist())
                        if iou > best_iou:
                            best_iou = iou
                            best_gt = j

                    if best_iou >= iou_threshold and best_gt not in matched:
                        TP += 1
                        matched.add(best_gt)
                    else:
                        FP += 1

                FN += len(gt_boxes) - len(matched)

    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)
    f1 = 2 * precision * recall / (precision + recall + 1e-6)

    # simplified mAP
    map50 = precision * recall

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"mAP@0.5: {map50:.4f}")

    return precision, recall, f1, map50
```

```python
import os
import torch
from PIL import Image
import torchvision.transforms as T

class PotholeDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms
        self.imgs = sorted(os.listdir(os.path.join(root, "images")))
        self.labels = sorted(os.listdir(os.path.join(root, "labels")))

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "images", self.imgs[idx])
        label_path = os.path.join(self.root, "labels", self.labels[idx])

        img = Image.open(img_path).convert("RGB")
        w, h = img.size

        boxes = []
        labels = []

        with open(label_path) as f:
            for line in f.readlines():
                cls, x, y, bw, bh = map(float, line.strip().split())

                x1 = (x - bw/2) * w
                y1 = (y - bh/2) * h
                x2 = (x + bw/2) * w
                y2 = (y + bh/2) * h

                boxes.append([x1, y1, x2, y2])
                labels.append(1)

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)

        target = {"boxes": boxes, "labels": labels}

        if self.transforms:
            img = self.transforms(img)

        return img, target
```

```python
val_dataset = PotholeDataset(
    "/content/drive/MyDrive/dataset_final/valid",
    get_transform()
)

val_loader = torch.utils.data.DataLoader(
    val_dataset,
    batch_size=4,
    shuffle=False,
    collate_fn=lambda x: tuple(zip(*x))
)
```

```python
def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter = max(0, x2 - x1) * max(0, y2 - y1)

    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union = area1 + area2 - inter

    return inter / union if union > 0 else 0
```

```python
import torch
import torchvision
from torchvision.models.detection import retinanet_resnet50_fpn
from torchvision.models.detection.retinanet import RetinaNetClassificationHead

all_scores = []
all_matches = []

iou_threshold = 0.5

# Re-create and load the RetinaNet model
device = torch.device("cuda") # Ensure device is defined

model = retinanet_resnet50_fpn(weights="DEFAULT")

num_classes = 2
in_channels = model.backbone.out_channels
num_anchors = model.head.classification_head.num_anchors

model.head.classification_head = RetinaNetClassificationHead(
    in_channels,
    num_anchors,
    num_classes
)

# load weights from drive
model.load_state_dict(torch.load("/content/drive/MyDrive/retinanet_best.pth"))

model.to(device)
model.eval()

with torch.no_grad():
    for images, targets in val_loader:
        images = [img.to(device) for img in images]
        outputs = model(images)

        for i in range(len(outputs)):
            pred_boxes = outputs[i]['boxes'].cpu()
            scores = outputs[i]['scores'].cpu()
            gt_boxes = targets[i]['boxes'].cpu()

            for pb, score in zip(pred_boxes, scores):
                best_iou = 0

                for gb in gt_boxes:
                    iou = compute_iou(pb.tolist(), gb.tolist())
                    best_iou = max(best_iou, iou)

                match = 1 if best_iou >= iou_threshold else 0

                all_scores.append(score.item())
                all_matches.append(match)
```

```python
# ==============================
# 🔷 RetinaNet Confusion Matrix
# ==============================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 🔥 Count total ground truth boxes
num_gt = 0

for _, targets in val_loader:
    for t in targets:
        num_gt += len(t['boxes'])

# 🔥 Build labels (1 = pothole, 0 = background)
y_pred = []
y_true = []

# TP + FP
for match in all_matches:
    if match == 1:
        y_pred.append(1)  # predicted pothole
        y_true.append(1)  # actual pothole
    else:
        y_pred.append(1)  # predicted pothole
        y_true.append(0)  # false positive

# FN
num_tp = sum(all_matches)
num_fn = num_gt - num_tp

for _ in range(num_fn):
    y_pred.append(0)  # predicted background
    y_true.append(1)  # missed pothole

# 🔥 Confusion Matrix (POTHOLE FIRST)
cm = confusion_matrix(y_true, y_pred, labels=[1, 0])

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Pothole", "Background"]
)

disp.plot(cmap='Blues')

plt.title("RetinaNet Confusion Matrix")
plt.show()
```

```python
import numpy as np

all_scores_np = np.array(all_scores)
all_matches_np = np.array(all_matches)

threshold = 0.5

preds = all_scores_np >= threshold

TP = np.sum((preds == 1) & (all_matches_np == 1))
FP = np.sum((preds == 1) & (all_matches_np == 0))
FN = np.sum((preds == 0) & (all_matches_np == 1))

# Metrics
precision = TP / (TP + FP + 1e-6)
recall = TP / (TP + FN + 1e-6)
f1 = 2 * (precision * recall) / (precision + recall + 1e-6)
accuracy = TP / (TP + FP + FN + 1e-6)

# mAP@0.5 (approximation)
map50 = precision * recall

print("RetinaNet Metrics:")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Accuracy: {accuracy:.4f}")
print(f"mAP@0.5 (approx): {map50:.4f}")
```

```python
import torch
import torchvision
import os
from PIL import Image
import torchvision.transforms as T
import matplotlib.pyplot as plt
import numpy as np
```

```python
class PotholeDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms=None):
        self.root = root
        self.transforms = transforms
        self.imgs = sorted(os.listdir(os.path.join(root, "images")))
        self.labels = sorted(os.listdir(os.path.join(root, "labels")))

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root, "images", self.imgs[idx])
        label_path = os.path.join(self.root, "labels", self.labels[idx])

        img = Image.open(img_path).convert("RGB")
        w, h = img.size

        boxes = []
        labels = []

        with open(label_path) as f:
            for line in f.readlines():
                cls, x, y, bw, bh = map(float, line.strip().split())

                x1 = (x - bw/2) * w
                y1 = (y - bh/2) * h
                x2 = (x + bw/2) * w
                y2 = (y + bh/2) * h

                boxes.append([x1, y1, x2, y2])
                labels.append(1)

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.int64)

        target = {"boxes": boxes, "labels": labels}

        if self.transforms:
            img = self.transforms(img)

        return img, target
```

```python
def get_transform():
    return T.Compose([T.ToTensor()])

val_dataset = PotholeDataset(
    "/content/drive/MyDrive/dataset_final/valid",
    get_transform()
)

val_loader = torch.utils.data.DataLoader(
    val_dataset,
    batch_size=4,
    shuffle=False,
    collate_fn=lambda x: tuple(zip(*x))
)
```

```python
def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter = max(0, x2 - x1) * max(0, y2 - y1)

    area1 = (box1[2]-box1[0]) * (box1[3]-box1[1])
    area2 = (box2[2]-box2[0]) * (box2[3]-box2[1])

    union = area1 + area2 - inter

    return inter / union if union > 0 else 0
```

```python
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor

device = torch.device("cuda")

model = fasterrcnn_resnet50_fpn(weights="DEFAULT")

num_classes = 2
in_features = model.roi_heads.box_predictor.cls_score.in_features

model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

# LOAD YOUR MODEL
model.load_state_dict(torch.load("/content/drive/MyDrive/fasterrcnn_best.pth"))

model.to(device)
model.eval()
```

```python
all_scores = []
all_matches = []

iou_threshold = 0.5

with torch.no_grad():
    for images, targets in val_loader:
        images = [img.to(device) for img in images]
        outputs = model(images)

        for i in range(len(outputs)):
            pred_boxes = outputs[i]['boxes'].cpu()
            scores = outputs[i]['scores'].cpu()
            gt_boxes = targets[i]['boxes'].cpu()

            for pb, score in zip(pred_boxes, scores):
                best_iou = 0

                for gb in gt_boxes:
                    iou = compute_iou(pb.tolist(), gb.tolist())
                    best_iou = max(best_iou, iou)

                match = 1 if best_iou >= iou_threshold else 0

                all_scores.append(score.item())
                all_matches.append(match)
```

```python
import numpy as np

# Convert to numpy
scores = np.array(all_scores)
matches = np.array(all_matches)

# Sort by confidence (descending)
sorted_indices = np.argsort(-scores)
scores = scores[sorted_indices]
matches = matches[sorted_indices]

# True Positives and False Positives
tp = matches
fp = 1 - matches

# Cumulative sums
tp_cumsum = np.cumsum(tp)
fp_cumsum = np.cumsum(fp)

# Total ground truth boxes
total_gt = sum([len(t['boxes']) for _, t in val_loader.dataset])

# Precision and Recall
precision = tp_cumsum / (tp_cumsum + fp_cumsum + 1e-6)
recall = tp_cumsum / (total_gt + 1e-6)

# ---- AP Calculation (COCO-style interpolation) ----

# Append boundary points
precision = np.concatenate(([0.0], precision, [0.0]))
recall = np.concatenate(([0.0], recall, [1.0]))

# Make precision monotonically decreasing
for i in range(len(precision) - 1, 0, -1):
    precision[i-1] = max(precision[i-1], precision[i])

# Find points where recall changes
indices = np.where(recall[1:] != recall[:-1])[0]

# Compute AP
ap = np.sum((recall[indices + 1] - recall[indices]) * precision[indices + 1])

print(f"mAP@0.5 (Faster R-CNN): {ap:.4f}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

all_matches = []
all_scores = []

iou_threshold = 0.5
num_gt = 0  # total ground truth boxes

def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter = max(0, x2 - x1) * max(0, y2 - y1)

    area1 = (box1[2] - box1[0]) * (box1[3] - box1[1])
    area2 = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union = area1 + area2 - inter

    return inter / union if union > 0 else 0


with torch.no_grad():
    for images, targets in val_loader:
        images = [img.to(device) for img in images]
        outputs = model(images)

        for i in range(len(outputs)):
            pred_boxes = outputs[i]['boxes'].cpu()
            scores = outputs[i]['scores'].cpu()
            gt_boxes = targets[i]['boxes'].cpu()

            num_gt += len(gt_boxes)

            for pb, score in zip(pred_boxes, scores):
                best_iou = 0

                for gb in gt_boxes:
                    iou = compute_iou(pb.tolist(), gb.tolist())
                    best_iou = max(best_iou, iou)

                match = 1 if best_iou >= iou_threshold else 0

                all_matches.append(match)
                all_scores.append(score.item())


# 🔥 Build labels (1 = pothole, 0 = background)
y_pred = []
y_true = []

# TP + FP
for match in all_matches:
    if match == 1:
        y_pred.append(1)  # predicted pothole
        y_true.append(1)  # actual pothole
    else:
        y_pred.append(1)  # predicted pothole
        y_true.append(0)  # actually background (FP)

# FN
num_tp = sum(all_matches)
num_fn = num_gt - num_tp

for _ in range(num_fn):
    y_pred.append(0)  # predicted background
    y_true.append(1)  # actual pothole


# 🔥 CONFUSION MATRIX (POTHOLE FIRST)
cm = confusion_matrix(y_true, y_pred, labels=[1, 0])

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Pothole", "Background"]
)

disp.plot(cmap='Blues')

plt.title("Faster R-CNN Confusion Matrix")
plt.show()
```

```python
import numpy as np

all_scores = np.array(all_scores)
all_matches = np.array(all_matches)

threshold = 0.5  # confidence threshold

predictions = all_scores >= threshold

TP = np.sum((predictions == 1) & (all_matches == 1))
FP = np.sum((predictions == 1) & (all_matches == 0))
FN = np.sum((predictions == 0) & (all_matches == 1))

precision = TP / (TP + FP + 1e-6)
recall = TP / (TP + FN + 1e-6)
f1 = 2 * (precision * recall) / (precision + recall + 1e-6)
accuracy = TP / (TP + FP + FN + 1e-6)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
thresholds = np.linspace(0.1, 0.9, 20)

precisions = []
recalls = []

for t in thresholds:
    TP = FP = FN = 0

    for score, match in zip(all_scores, all_matches):
        if score >= t:
            if match == 1:
                TP += 1
            else:
                FP += 1
        else:
            if match == 1:
                FN += 1

    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)

    precisions.append(precision)
    recalls.append(recall)

plt.figure()
plt.plot(recalls, precisions)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curve (Faster R-CNN)")
plt.show()
```

CROSS MODEL EVALUATION

```python
def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter_area = max(0, x2 - x1) * max(0, y2 - y1)

    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])

    union_area = box1_area + box2_area - inter_area + 1e-6

    return inter_area / union_area
```

```python
import torch

def evaluate_model(model, dataloader, device, iou_thresh=0.5, score_thresh=0.5):
    model.eval()

    TP, FP, FN = 0, 0, 0

    with torch.no_grad():
        for images, targets in dataloader:
            images = [img.to(device) for img in images]
            outputs = model(images)

            for i in range(len(outputs)):
                pred_boxes = outputs[i]['boxes'].cpu()
                scores = outputs[i]['scores'].cpu()
                gt_boxes = targets[i]['boxes'].cpu()

                matched = set()

                for pb, score in zip(pred_boxes, scores):
                    if score < score_thresh:
                        continue

                    best_iou = 0
                    best_gt = -1

                    for j, gb in enumerate(gt_boxes):
                        iou = compute_iou(pb.tolist(), gb.tolist())
                        if iou > best_iou:
                            best_iou = iou
                            best_gt = j

                    if best_iou >= iou_thresh and best_gt not in matched:
                        TP += 1
                        matched.add(best_gt)
                    else:
                        FP += 1

                FN += len(gt_boxes) - len(matched)

    # Metrics
    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)
    f1 = 2 * (precision * recall) / (precision + recall + 1e-6)
    accuracy = TP / (TP + FP + FN + 1e-6)

    return precision, recall, f1, accuracy
```
